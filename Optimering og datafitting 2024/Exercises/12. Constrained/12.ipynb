```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)

# Define constraint functions
constraint1 = 1 - X1 - X2
constraint2 = 1 - X1 + X2
constraint3 = 1 + X1 - X2
constraint4 = 1 + X1 + X2

# Create the feasible region
feasible = (constraint1 >= 0) & (constraint2 >= 0) & (constraint3 >= 0) & (constraint4 >= 0)

# Define the objective function
def objective_function(x1, x2, t):
    return (x1 - 3/2) ** 2 + (x2 - t) ** 4

# Plotting function
def plot_feasible_and_contours(t_values):
    plt.figure(figsize=(10, 10))
    
    # Plot feasible region
    plt.contourf(X1, X2, feasible, levels=[0, 1], colors=['#d3d3d3'], alpha=0.5)
    plt.title("Feasible Set and Contours of Objective Function")
    
    # Plot contours for each t value
    for t in t_values:
        Z = objective_function(X1, X2, t)
        contours = plt.contour(X1, X2, Z, levels=10, alpha=0.8, linewidths=1.5, label=f"t={t}")
        plt.clabel(contours, inline=True, fontsize=8)
    
    # Constraint lines
    plt.plot(x1, -1-x1, 'r--', label=r'$1 - x_1 - x_2 = 0$')
    plt.plot(x1, 1+x1, 'g--', label=r'$1 - x_1 + x_2 = 0$')
    plt.plot(x1, 1-x1, 'b--', label=r'$1 + x_1 - x_2 = 0$')
    plt.plot(x1, -1+x1, 'm--', label=r'$1 + x_1 + x_2 = 0$')
    
    # Labels and legend
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Set different values of t and plot
t_values = [-1, 0, 0.5, 1]
plot_feasible_and_contours(t_values)
```

```python
# Re-import necessary libraries after environment reset
import numpy as np
import matplotlib.pyplot as plt

# Define the feasible set (unit circle)
theta = np.linspace(0, 2 * np.pi, 500)
x1_circle = np.cos(theta)
x2_circle = np.sin(theta)

# Define the objective function
def objective_function(x1, x2):
    return x1 * x2

# Grid for contour plot
x1 = np.linspace(-1.5, 1.5, 400)
x2 = np.linspace(-1.5, 1.5, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = objective_function(X1, X2)

# Plotting
plt.figure(figsize=(8, 8))
# Plot the feasible set (unit circle)
plt.plot(x1_circle, x2_circle, 'r-', label='Unit Circle (Feasible Set)')

# Contour plot of the objective function
contours = plt.contour(X1, X2, Z, levels=20, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)

# Labels and grid
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Feasible Set and Contour Plot of $f(x_1, x_2) = x_1 x_2$')
plt.legend()
plt.grid(alpha=0.3)
plt.axis('equal')
plt.show()
```
