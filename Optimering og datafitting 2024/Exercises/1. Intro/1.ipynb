1.1 Convexity

```python
def f(x):
    return x - np.log(x)

def df(x):
    return 1 - 1/x

def d2f(x):
    return 1/(x**2)
```

```python
x = np.arange(0.01,2.01,0.01)

y = f(x)
y_prime = df(x)
y_double_prime = d2f(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, label="f(x)")
ax.plot(x, y_prime, label="df(x)")
ax.plot(x, y_double_prime, label="d2f(x)")

ax.grid(True)
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Plot of f(x), f'(x), and f''(x)")
ax.set_ylim(-5,5)
ax.set_xlim(0,2)

plt.show()
```

You can tell by the plot of f(x) that it is convex, since is fulfills the inequality from Definition 1.

Since d2f(x) always is positive, it is convex.

```python
# minimum ved:
# df(x) 1 - 1/x = 0 -> x = 1
# This is the only solution (unique)

x_star = 1
f_min = f(x_star)

# It is a minimum if d2f(x_star) > 0
is_minimum = d2f(x_star) > 0

x_star, f_min, is_minimum
```
