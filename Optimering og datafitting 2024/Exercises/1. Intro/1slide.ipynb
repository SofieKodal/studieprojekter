```python
import numpy as np
import scipy
from scipy.linalg import lu
import matplotlib
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(-5, 5, 0.005)
x2 = x1
X, Y = np.meshgrid(x1, x2)
F = 3*X**2+2*Y**2+X*Y+3*X+2*Y+4
fig, ax=plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, F, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
plt.show()

fig, ax = plt.subplots()
v = np.concatenate((np.arange(0, 10, 2), np.arange(10, 100, 5), np.arange(100, 200, 20)), axis=None)
c = ax.contour(X, Y, F, v, linewidths=2)
norm = matplotlib.colors.Normalize(vmin=c.cvalues.min(), vmax=c.cvalues.max())
sm = matplotlib.cm.ScalarMappable(norm=norm, cmap=c.cmap)
fig.colorbar(sm, ax=ax)
plt.show()
```
